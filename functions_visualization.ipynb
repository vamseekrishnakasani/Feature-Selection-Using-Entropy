```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def line_func(x, a, b):
    return a*x + b
```

```python
def sigmoid_func(x):
    return 1/(1+np.exp(-x))
```

```python
def tanh_func(x):
    return np.tanh(x)
```

```python
def relu_func(x):
    return np.maximum(0, x)
```

```python
def leaky_relu_func(x):
    return np.maximum(0.01*x, x)
```

```python
def softmax_func(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.linspace(-10, 10, 100)
```

```python
# Plotting
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Activation Functions')

# Flatten the 2D array of axes for easier access
ax = ax.ravel()

ax[0].plot(x, line_func(x, 2, 3), c='r')
ax[0].set_title('Linear')

ax[1].plot(x, sigmoid_func(x), c='r')
ax[1].set_title('Sigmoid')

ax[2].plot(x, tanh_func(x), c='r')
ax[2].set_title('Tanh')

ax[3].plot(x, relu_func(x), c='r')
ax[3].set_title('ReLU')

ax[4].plot(x, leaky_relu_func(x), c='r')
ax[4].set_title('Leaky ReLU')

ax[5].plot(x, softmax_func(x), c='r')
ax[5].set_title('Softmax')

plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()
```

```python
# Importing libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns 
matplotlib.style.use('fivethirtyeight')

# data
x = pd.DataFrame({
    # Distribution with lower outliers
    'x1': np.concatenate([np.random.normal(20, 2, 1000), np.random.normal(1, 2, 25)]),
    # Distribution with higher outliers
    'x2': np.concatenate([np.random.normal(30, 2, 1000), np.random.normal(50, 2, 25)]),
})
np.random.normal

scaler = preprocessing.RobustScaler()
robust_df = scaler.fit_transform(x)
robust_df = pd.DataFrame(robust_df, columns =['x1', 'x2'])

scaler = preprocessing.StandardScaler()
standard_df = scaler.fit_transform(x)
standard_df = pd.DataFrame(standard_df, columns =['x1', 'x2'])

scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(x)
minmax_df = pd.DataFrame(minmax_df, columns =['x1', 'x2'])

fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols = 4, figsize =(20, 5))

ax1.set_title('Before Scaling')
sns.kdeplot(x['x1'], ax = ax1, color ='r')
sns.kdeplot(x['x2'], ax = ax1, color ='b')

ax2.set_title('After Robust Scaling')
sns.kdeplot(robust_df['x1'], ax = ax2, color ='red')
sns.kdeplot(robust_df['x2'], ax = ax2, color ='blue')

ax3.set_title('After Standard Scaling')
sns.kdeplot(standard_df['x1'], ax = ax3, color ='black')
sns.kdeplot(standard_df['x2'], ax = ax3, color ='g')

ax4.set_title('After Min-Max Scaling')
sns.kdeplot(minmax_df['x1'], ax = ax4, color ='black')
sns.kdeplot(minmax_df['x2'], ax = ax4, color ='g')
plt.show()
```

```python
import tensorflow as tf

@tf.function
def compute(x, y):
    return x * y + y

x = tf.constant(2.0)
y = tf.constant(3.0)

# Create a writer to log the graph
writer = tf.summary.create_file_writer("./logs")

# Start the profiler
tf.profiler.experimental.start('./logs')  # Start TensorFlow Profiler

# Record the graph in TensorBoard
with writer.as_default():
    tf.summary.trace_on(graph=True, profiler=True)  # Start tracing the graph
    compute(x, y)  # Run the function to trace it
    tf.summary.trace_export(name="compute_graph", step=0, profiler_outdir="./logs")  # Save the trace

# Stop the profiler
tf.profiler.experimental.stop()  # Stop TensorFlow Profiler
```

```python
import tensorflow as tf

# Define a simple Keras model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(10,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

@tf.function
def train_step(x):
    return model(x)

# Generate sample input data
x = tf.random.normal((100, 10))

# Create a writer for TensorBoard logs
writer = tf.summary.create_file_writer("./logs")

# Clear previous traces
tf.summary.trace_off()

tf.summary.trace_on(graph=True, profiler=True, profiler_outdir="./logs")
train_step(x)  # Run the function to trace the graph

# Trace the graph with the correct profiler output directory
with writer.as_default():
    tf.summary.trace_export(name="keras_model_graph", step=0, profiler_outdir="./logs")
```

```python
import tensorflow as tf
import numpy as np

x = np.array([np.random.randint(10) for _ in range(100)]).reshape(-1, 1)
print(x.shape)
y = np.array(list(map(lambda i: i * 2, x)))
print(y.shape)


class MyModel(tf.keras.Model):
    def __init__(self):
        super(MyModel, self).__init__()
        self.dense = tf.keras.layers.Dense(1)

    def call(self, x):
        return self.dense(x)

# Create an instance of the model
model = MyModel()
model.summary()

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x, y, epochs=100)
```

```python
import tensorflow as tf

class MyModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.weight = tf.Variable(0.5, trainable=True)  # Custom trainable variable

    def __call__(self, x):
        return self.weight * x  # Simple linear operation

# Create and use the module
module = MyModule()
output = module(tf.constant(2.0))
print("Output:", output.numpy())  # Output: 1.0 (0.5 * 2.0)
```
